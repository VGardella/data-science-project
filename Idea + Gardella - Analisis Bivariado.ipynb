```python
#Lo primero que debemos hacer es cargar la base de datos a usar:

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Bases de Datos/Clasificación de clientes de banco - Train.csv')
```

```python
#Vamos a graficar la relacion entre varios atributos:

#Prestamos personales vs. Puesto de Trabajo.

sns.set()
fig, ax = plt.subplots(figsize=(16,8))
trabajo_prestamo = sns.histplot(data = df, x = df["job"], hue = df["loan"], multiple = "dodge", shrink=.8, edgecolor = "white")
ax.set(xlabel = "Trabajo", ylabel = "Numero de Clientes", title = "Relación: Clientes con Prestamos Personales según su Puesto de Trabajo.")
plt.xticks(rotation=45)
```

Podemos ver que la gran mayoria de los clientes que sacaron prestamos personales tienen trabajos administrativos, técnicos o del tipo "blue collar"(es decir, trabajadores de jerarquía empresarial baja u obreros), pudiendo extrapolar que la mayor cantidad de prestamos los sacan trabajadores de clase socioeconómica baja-media.

```python
#Edad con Presatmos Hipotecarios (o prestamos para comprar una casa):

sns.set()
fig, ax = plt.subplots(figsize=(16,8))
edad_hipot = sns.histplot(data=df, x=df["age"], binwidth=5, hue=df["housing"], palette=sns.color_palette("Set1", 3), multiple="stack")
ax.set(xlabel = "Edad", ylabel = "Número de clientes", title="Número de clientes con prestamos hipotecarios segun su edad")
```

Podemos ver una cantidad similar de clientes con y sin prestamos hipotecarios, siendo principalmente gente de entre 25-40 años quienes poseen prestamos a su nombre.

```python
#Relaciones Nivel educatico vs tenencia de Prestamos Hipotecarios:

sns.set()
fig, ax = plt.subplots(figsize=(16,8))
trabajo_prestamo = sns.histplot(data = df, x = df["education"], hue = df["housing"], multiple = "dodge", shrink=.5, edgecolor = "grey")
ax.set(xlabel = "Nivel educativo", ylabel = "Cantidad de Clientes", title = "Clientes con Prestamo Hipotecario según nivel educativo.")
plt.xticks(rotation=45)
```

```python
#Relación entre Nivel educativo y Tipo de trabajo de los clientes:

sns.set()
fig, ax = plt.subplots(figsize=(16,8))
trabajo_prestamo = sns.histplot(data = df, x = df["education"], hue = df["job"], multiple = "dodge", shrink=.5, edgecolor = "grey")
ax.set(xlabel = "Nivel Educativo", ylabel = "Cantidad de clientes", title = "Clientes con trabajo según nivel educativo.")
plt.xticks(rotation=45)
```

```python
# Relación entre la duración de la llamda y la suscripción a un deposito de largo plazo

fig, ax = plt.subplots(figsize=(16,8))
sns.boxplot(x = 'y', y = 'duration', data = df)
ax.set(xlabel = 'Suscripción a depósito de largo plazo', ylabel = 'Duración de la llamada', title = 'Distribución de la duración de la llamada vs Suscripción a deposito de largo plazo')
plt.show()
```

En el gráfico de cajas de la duración de la última llamda se puede observar que el grueso de las personas que mantuvieron llamadas de menor duración, tuvieron un menor nivel de aceptación de la propuesta que aquellas que mantuvieron llamadas de mayor duración. Así mismo, los valores extremos tienen un alto nivel de superposición.

```python
# Distribución de la duración de la última llamada y porcentaje de suscripción a deposito a largo plazo

fig, ax = plt.subplots(figsize=(16,8))
sns.histplot(data = df, x = df['duration'], binwidth = 60, hue = df['y'], multiple = 'stack',edgecolor = 'grey')
ax.set(xlabel = 'Duración de la llamada', ylabel = 'Cantidad de clientes', title = 'Duración de la llamda vs Suscripción depósito largo plazo')
plt.show()
```

En este gráfico se puede observar que la distribución de la duración de la ultima llamada no es normal. Esto explica el sesgo y el alto número de valores extremos que se observan en el gráfico de cajas.
